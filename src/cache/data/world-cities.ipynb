```python
import pandas as pd
import json
```

```python
def namespace(row):
    city = row['city'].replace(' ', '-').lower()
    country = row['country'].replace(' ', '-').lower()
    return f"{city}-{country}"
```

```python
world_cities = pd.read_csv("world-cities.csv")
world_cities = world_cities.fillna("")
world_cities["processed"] = False
world_cities['namespace'] = world_cities.apply(namespace, axis=1)
world_cities = world_cities[['city', 'state', 'country', 'namespace', 'lat', 'lng', 'population', 'processed']]
world_cities
```

```python
city_json = []

for row in world_cities.iterrows():
    city_json.append(row[1].to_dict())

with open("../cache/cache.json", "w") as file:
    json.dump(city_json, file, indent=4)
```
